```python
# 다중 그래프, range(5): 0 ~ 4
d = np.arange(0., 10., 1.) # range 기능 확장, 시작값, 종료값, 증가값
print(d)
print(type(d))
# x, y, 선의 스타일(-: 실선, --: 점선 및 색깔(rgb등) 나열
# 예) x, y, 선 스타일, x, y, 선 스타일, x, y, 선 스타일...
# r-: 빨간색 실선, g-: 초록색 실선, b--: 파란색 점선
plt.plot(d, d*2, 'r-', d, d*3, 'g-', d, d*4, 'b--')
plt.title('X Y 그래프')   # 그래프 제목
plt.xlabel('X 값')
plt.ylabel('Y 값')
plt.legend(('상품1', '상품2', '상품3'))             # 범례 출력
plt.show()
```

```python
d = np.arange(0., 10., 1.) # range 기능 확장, 시작값, 종료값, 증가값

plt.figure(figsize=(8, 5)) # 단위: 인치

# 서브 그래프 1
plt.subplot(2, 2, 1)     # 가로방향 2개, 새로방향 2개, 1: 첫번째 그래프
plt.plot(d, d*2, 'r-', label='상품1')     # x, y, 빨간색 실선
plt.ylabel('Y 값')
plt.legend()              # 범례 출력
plt.title('X Y 그래프')   # 그래프 제목

# 서브 그래프 2
plt.subplot(2 , 2, 2)    #: 가로방향 2개, 새로방향 2개, 2: 두번째 그래프
plt.plot(d, d*-2, 'g--', label='상품2')  # x, y, 초록색 점선
plt.legend()              # 범례 출력
plt.title('X Y 그래프')   # 그래프 제목

# 서브 그래프 3
plt.subplot(2, 2, 3)      # 가로방향 2개, 새로방향 2개, 세번째 그래프
plt.plot(d, d*-2, 'b--', label='상품3')  # x, y, 파란색 점선
plt.xlabel('X 값')
plt.ylabel('Y 값')
plt.legend()               # 범례 출력

# 서브 그래프 4
plt.subplot(2, 2, 4)      # 가로방향 2개, 새로방향 2개, 네번째 그래프
plt.plot(d, d*2, 'y-', label='상품4')     # x, y, 노란색 실선
plt.xlabel('X 값')
plt.ylabel('Y 값')
plt.legend()               # 범례 출력

plt.show()
```

```python
def f2(x, w):
    return (x - w) * x * (x + 2) 

plt.figure(figsize=(10, 3)) # (A) figure 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) # (B) 그래프의 간격을 지정
for i in range(6): # 0 ~ 5
    plt.subplot(2, 3, i + 1) # 2행 3열중에 출력될 그래프의 위치 지정
    plt.title(i + 1)
    plt.plot(x, f2(x, i))
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

### Vector 데이터의 이미지로 출력 

```python
y = np.zeros((10, 10)) # 숫자 0을 이용하여 10 x 10 배열 생성 
y
```

```python
for i in range(0, 10, 1): # row: 0 ~ 9
    for j in range(9, -1, -1): # col: 9 ~ 0
        if (i == (9-j)):
#            y[i, j] = 1  # y[0, 9] = 1, y[1, 8] = 1, y[2, 7] = 1
             y[i, j] = 1 * (i * 0.1)
print(y)    
```

```python
plt.figure(figsize=(3.5, 3))
plt.gray()
plt.pcolor(y) # 2차원 배열을 색상으로 지정, Gray: 0-black, 1-white
plt.colorbar()
plt.show()
```

```python
y = np.zeros((100, 100)) # 숫자 0을 이용하여 100 x 100 배열 생성 
y
```

```python
for i in range(0, 100, 1): # row: 0 ~ 99
    for j in range(99, -1, -1): # col: 99 ~ 0
        if (i == (99-j)):
#            y[i, j] = 1  # y[0, 9] = 1, y[1, 8] = 1, y[2, 7] = 1
             y[i, j] = (i/10) * 0.1
print(y)    
```

```python
plt.figure(figsize=(3.5, 3))
plt.gray()
plt.pcolor(y) # 2차원 배열을 색상으로 지정, Gray: 0-black, 1-white
plt.colorbar()
plt.show()
```

### boxplot

```python
import matplotlib.pyplot as plt
import random
import numpy as np

result = []
for i in range(10) : # 0 - 9
    result.append(random.randint(1,100)) # 1 ~ 10
print(sorted(result))

result = np.array(result) # ndarray
print("1/4: " + str(np.percentile(result,25)))
print("2/4: " + str(np.percentile(result,50)))
print("3/4: " + str(np.percentile(result,75)))
plt.style.use('ggplot')
plt.boxplot(result)
plt.show()
# 이상치가 없음.
```

```python
x = [1, 2, 3, 4, 5]
acc = [0.8372093, 0.83333333, 0.75609756, 0.87804878, 0.80487805]

plt.rcParams["font.size"] = 12         # 글자 크기
plt.rcParams["figure.figsize"] = (8, 4) # 10:4의 그래프 비율, subplot 사용시 적용
# plt.figure(figsize=(20, 5))            # subplot 사용시 정상작동안함.
fig = plt.figure()
ax = fig.add_subplot(111) #  fig.add_subplot(1, 1, 1) # 1행 1열 첫번째 plot

plt.plot(x, acc, c='g')
plt.ylim(0, 1)
plt.xlabel('학습 횟수')
plt.ylabel('정확도')

# zip([1, 2, 3, 4, 5], [0.8372093  0.83333333 0.75609756 0.87804878 0.80487805])
# (1, 0.8372093), (2, 0.83333333)...
for xy in zip(x, acc):
    # print(type(xy))  # <class 'tuple'>
    # 출력 문자열, 기준 xy 지점, xytext: 문자열 출력 위치
    ax.annotate('%.2f' % xy[1], xy=xy, xytext=xy)  
plt.grid(True)
plt.show()
```
