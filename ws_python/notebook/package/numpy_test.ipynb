### 통계 기초

```python
x = [1, 2, 3, 4, 5]
errors = []    # 평균에 대한 편차
errors2 = []  # 평균에 대한 편차의 제곱

total = 0
# for i in range(len(x)):
#     total = total + x[i]

for i in x:
    total = total + i
print('total:',total)    

avg = total / len(x)
print('avg:',avg)
```

```python
print('평균에 대한 편차')
for i in x:
    errors.append(avg - i)
```

```python
for i in errors:
    print('%.1f' % i)
```

```python
print('편차의 합')
errors_tot = 0
for i in errors:
    errors_tot = errors_tot + i
print(errors_tot)
```

```python
print('평균에 대한 편차의 제곱')
for i in errors:
    errors2.append(i ** 2)
```

```python
for i in errors2:
    print(i)
```

```python
errors2=[]
for i in x:
    errors2.append((avg-i) ** 2) # (평균 - 변량) ** 2
```

```python
for i in errors2:
    print(i)
```

```python
print('편차의 제곱의 합')
errors_tot2 = 0
for i in errors2:
    errors_tot2 = errors_tot2 + i
    
print(errors_tot2)
```

```python
print("분산")  # 편차의 제곱의 합 / 변량의 수
variance = errors_tot2 / len(x);
print(variance)
```

```python
print(np.sqrt(0)) # 현재의 수를 만들기위해 제곱 해야하는 수
print(np.sqrt(1))
print(np.sqrt(2))
print(1.4142135623730951 * 1.4142135623730951)
print(np.sqrt(3))
print(1.7320508075688772 * 1.7320508075688772)
print(np.sqrt(4))
print(np.sqrt(9))
print(np.sqrt(16))
```

```python
print('표준편차, 평균 제곱근 오차')
sd = np.sqrt(variance) 
print(sd)
```

```python
def f(data):
    m = np.mean(data)
    v = np.var(data)
    s = np.std(data)
    
    return m, v, s

x = [1, 2, 3, 4, 5]    
m, v, s = f(x)
print('mean:', m , ' variance:', v, 'standard diviation:', s)
```
