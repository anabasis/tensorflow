```python
import numpy as np

print('np.pi:', np.pi)
print(np.arange(5))
signal_data = np.cos(np.arange(1600)*(20*np.pi/1000))[:,None]  # COS 데이터 1600개 생성
print(type(signal_data))
print(signal_data.shape)
print(signal_data[0:10])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plot_x = np.arange(1600)
plot_y = signal_data
plt.plot(plot_x, plot_y)
plt.show()
```
